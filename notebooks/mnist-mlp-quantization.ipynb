# MNIST MLP 가중치 양자화 실습

파이토치로 MNIST를 MLP로 간단히 학습시킨 뒤, 학습된 **가중치의 분포**를 들여다보고 두 가지 양자화 방법을 직접 적용해 봅니다.

1. **가중치 분포 히스토그램** — 신경망 가중치가 정말 0 근처에 몰려 있는지 눈으로 확인
2. **방법 A — K-Means 양자화** (Deep Compression, Han et al. 2016): 비슷한 값을 대표값으로 묶기 (비균등)
3. **방법 B — Linear 양자화** (Jacob et al. 2018): 아핀 매핑 `r = S(q - Z)`로 균등 격자에 매핑
4. **두 방법 비교** — 같은 비트 수에서 양자화 오차와 정확도

> 관련 블로그 글: [K-Means 기반 양자화](https://ho4040.github.io/ko/posts/kmeans-weight-quantization/) · [Linear Quantization](https://ho4040.github.io/ko/posts/linear-quantization/)

## 0. 준비

필요한 라이브러리는 Colab에 모두 기본 설치되어 있습니다 (torch, torchvision, scikit-learn, matplotlib).

```python
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
import torchvision
from torchvision import transforms
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

try:
    from safetensors.torch import save_file, load_file
except ImportError:
    import subprocess, sys as _sys
    subprocess.run([_sys.executable, "-m", "pip", "install", "-q", "safetensors"])
    from safetensors.torch import save_file, load_file

torch.manual_seed(0)
np.random.seed(0)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("device:", device)
```

## 1. MNIST를 MLP로 학습

작은 3층 MLP(784 → 256 → 128 → 10)를 2 에폭만 학습합니다. 목적은 최고 성능이 아니라 **양자화해 볼 가중치**를 얻는 것이므로 짧게 돌립니다.

```python
transform = transforms.Compose([transforms.ToTensor(),
                                transforms.Normalize((0.1307,), (0.3081,))])
train_ds = torchvision.datasets.MNIST("./data", train=True,  download=True, transform=transform)
test_ds  = torchvision.datasets.MNIST("./data", train=False, download=True, transform=transform)
train_loader = DataLoader(train_ds, batch_size=128, shuffle=True)
test_loader  = DataLoader(test_ds,  batch_size=256, shuffle=False)

class MLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(28*28, 256)
        self.fc2 = nn.Linear(256, 128)
        self.fc3 = nn.Linear(128, 10)
    def forward(self, x):
        x = x.view(x.size(0), -1)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        return self.fc3(x)

model = MLP().to(device)
opt = torch.optim.Adam(model.parameters(), lr=1e-3)

def accuracy(m):
    m.eval()
    correct = total = 0
    with torch.no_grad():
        for x, y in test_loader:
            x, y = x.to(device), y.to(device)
            pred = m(x).argmax(1)
            correct += (pred == y).sum().item()
            total += y.size(0)
    return correct / total

for epoch in range(2):
    model.train()
    for x, y in train_loader:
        x, y = x.to(device), y.to(device)
        opt.zero_grad()
        loss = F.cross_entropy(model(x), y)
        loss.backward()
        opt.step()
    print(f"epoch {epoch+1}  test acc = {accuracy(model):.4f}")

base_acc = accuracy(model)
print("baseline (float32) accuracy:", base_acc)

# 학습된 가중치를 safetensors 파일로 저장 (실전에서 다루는 체크포인트 형태)
save_file({k: v.cpu().contiguous() for k, v in model.state_dict().items()},
          "mnist_mlp.safetensors")
print("saved -> mnist_mlp.safetensors")
```

## 3. 가중치 분포 히스토그램

파일에서 불러온 첫 번째 층(`fc1`)의 가중치를 펼쳐 히스토그램을 그려봅니다. 신경망 가중치의 전형적인 특징 — **0 근처에 종 모양으로 몰려 있고 꼬리가 긴** 분포 — 를 확인할 수 있습니다. 이 성질이 이후 두 양자화 방법의 성능 차이를 만듭니다.

```python
w = model.fc1.weight.detach().cpu().numpy()
print("fc1.weight shape:", w.shape, " (총", w.size, "개 파라미터)")
print(f"min={w.min():.3f}  max={w.max():.3f}  mean={w.mean():.4f}  std={w.std():.3f}")

plt.figure(figsize=(7, 3.5))
plt.hist(w.reshape(-1), bins=200, color="#3b82f6")
plt.title("fc1 weight distribution (float32)")
plt.xlabel("weight value"); plt.ylabel("count")
plt.tight_layout(); plt.show()
```

## 3. 방법 A — K-Means 양자화 (Deep Compression)

비슷한 값의 가중치를 $k = 2^N$개의 클러스터로 묶고, 각 클러스터의 중심값(centroid)을 **코드북**으로 저장합니다. 각 가중치는 자기가 속한 클러스터의 대표값으로 복원됩니다.

- **비균등(non-uniform)**: 값이 몰린 곳(0 근처)에 대표값이 촘촘하게 배치됨
- 논문의 팁대로 초기값은 min~max를 균등 분할한 **선형 초기화**를 사용 (지역 최적해를 피함)

빠른 실행을 위해 코드북은 가중치의 부분표본으로 학습(fit)한 뒤 전체에 적용합니다.

```python
def kmeans_quantize(w, n_bits, fit_sample=20000):
    """가중치를 K-Means로 양자화. 복원된 가중치와 코드북, 인덱스를 반환."""
    shape = w.shape
    flat = w.reshape(-1, 1)
    k = 2 ** n_bits
    lo, hi = float(flat.min()), float(flat.max())
    init = np.linspace(lo, hi, k).reshape(-1, 1)          # 선형 초기화
    n = len(flat)
    idx = np.random.choice(n, min(fit_sample, n), replace=False)
    km = KMeans(n_clusters=k, init=init, n_init=1, max_iter=50).fit(flat[idx])
    codebook = km.cluster_centers_.reshape(-1)
    indices = km.predict(flat)
    recon = codebook[indices].reshape(shape)
    return recon, codebook, indices

N_BITS = 2
recon_km, codebook, indices = kmeans_quantize(w, N_BITS)
mse_km = ((w - recon_km) ** 2).mean()

# 저장 용량 비교 (인덱스 + 코드북)
num = w.size; k = 2 ** N_BITS
orig_bits = 32 * num
comp_bits = N_BITS * num + 32 * k
print(f"[K-Means] {N_BITS}-bit, k={k} clusters")
print("codebook (centroids):", np.sort(codebook).round(3))
print(f"quantization MSE = {mse_km:.6f}")
print(f"compression: {orig_bits/comp_bits:.2f}x  ({orig_bits//8} B -> {comp_bits//8} B)")

plt.figure(figsize=(7, 3.5))
plt.hist(w.reshape(-1), bins=200, alpha=0.4, color="#3b82f6", label="original")
plt.hist(recon_km.reshape(-1), bins=200, color="#e0a020", label="K-Means quantized")
plt.title(f"K-Means quantization ({N_BITS}-bit, {k} levels)")
plt.xlabel("weight value"); plt.ylabel("count"); plt.legend()
plt.tight_layout(); plt.show()
```

## 4. 방법 B — Linear 양자화 (아핀 매핑)

정수 $q$와 실수 $r$을 아핀 매핑으로 잇습니다.

$$r = S \cdot (q - Z), \qquad S = \frac{r_{\max}-r_{\min}}{q_{\max}-q_{\min}}, \qquad Z = \mathrm{round}\!\left(q_{\min} - \frac{r_{\min}}{S}\right)$$

- **균등(uniform)**: 대표값들이 일정한 간격($S$)으로 배치됨
- `symmetric=True`면 $Z=0$으로 고정 (가중치가 0 대칭일 때 유리, 정수 연산이 더 단순)

`S`, `Z`는 min/max로 계산하고, **양자화(round·clamp)와 복원(dequant)은 PyTorch 내장 연산** `torch.fake_quantize_per_tensor_affine`이 처리합니다. (임의 비트 폭을 `quant_min`/`quant_max`로 지정 가능)

```python
def linear_quantize(w, n_bits, symmetric=False):
    """PyTorch 내장 fake_quantize로 선형(아핀) 양자화. 복원값과 S, Z를 반환."""
    t = torch.as_tensor(w, dtype=torch.float32)
    if symmetric:
        qmax = 2 ** (n_bits - 1) - 1
        qmin = -qmax
        S = float(t.abs().max()) / qmax
        Z = 0
    else:
        qmin = -2 ** (n_bits - 1)
        qmax = 2 ** (n_bits - 1) - 1
        S = (float(t.max()) - float(t.min())) / (qmax - qmin)
        Z = int(round(qmin - float(t.min()) / S))
    # round/clamp/dequant 는 PyTorch가 처리 (가짜 양자화: 양자화 후 다시 float로 복원)
    recon = torch.fake_quantize_per_tensor_affine(t, S, Z, qmin, qmax)
    return recon.numpy(), S, Z

recon_lin, S, Z = linear_quantize(w, N_BITS, symmetric=False)
mse_lin = ((w - recon_lin) ** 2).mean()
print(f"[Linear] {N_BITS}-bit (asymmetric)")
print(f"scale S = {S:.5f}   zero point Z = {Z}")
print("levels:", np.unique(recon_lin.reshape(-1)).round(3))
print(f"quantization MSE = {mse_lin:.6f}")

plt.figure(figsize=(7, 3.5))
plt.hist(w.reshape(-1), bins=200, alpha=0.4, color="#3b82f6", label="original")
plt.hist(recon_lin.reshape(-1), bins=200, color="#22c55e", label="Linear quantized")
plt.title(f"Linear quantization ({N_BITS}-bit, uniform levels)")
plt.xlabel("weight value"); plt.ylabel("count"); plt.legend()
plt.tight_layout(); plt.show()
```

### 두 방법의 대표값 배치 비교

같은 2비트(4개 레벨)인데, **K-Means는 0 근처에 촘촘하게(비균등)**, **Linear는 일정 간격으로(균등)** 대표값을 놓습니다. 가중치가 0 근처에 몰려 있으므로 K-Means 쪽이 실제 값을 더 잘 맞춥니다.

```python
plt.figure(figsize=(7, 2.6))
plt.hist(w.reshape(-1), bins=200, alpha=0.3, color="#3b82f6", label="original")
for c in np.sort(codebook):
    plt.axvline(c, color="#e0a020", lw=1.5)
for lv in np.unique(recon_lin.reshape(-1)):
    plt.axvline(lv, color="#22c55e", lw=1.5, ls="--")
plt.axvline(np.nan, color="#e0a020", lw=1.5, label="K-Means levels")
plt.axvline(np.nan, color="#22c55e", lw=1.5, ls="--", label="Linear levels")
plt.title("representative levels: K-Means (solid) vs Linear (dashed)")
plt.xlabel("weight value"); plt.yticks([]); plt.legend(fontsize=8)
plt.tight_layout(); plt.show()

print(f"MSE  ---  K-Means: {mse_km:.6f}   Linear: {mse_lin:.6f}")
```

## 5. 비트 수에 따른 정확도 비교

두 방법으로 **모델의 모든 Linear 층 가중치**를 양자화한 뒤(복원값으로 교체), MNIST 테스트 정확도가 어떻게 변하는지 비트 수를 바꿔가며 비교합니다. 여기서는 가중치를 양자화 후 복원해 float로 추론하는 방식(fake quantization)으로 정확도 영향을 봅니다.

```python
import copy

def evaluate_with_quant(model, method, n_bits):
    m = copy.deepcopy(model)
    with torch.no_grad():
        for layer in [m.fc1, m.fc2, m.fc3]:
            wl = layer.weight.detach().cpu().numpy()
            if method == "kmeans":
                recon, *_ = kmeans_quantize(wl, n_bits)
            else:
                recon, *_ = linear_quantize(wl, n_bits, symmetric=True)
            layer.weight.copy_(torch.tensor(recon, dtype=torch.float32, device=device))
    return accuracy(m)

bits_list = [2, 3, 4, 8]
acc_km  = [evaluate_with_quant(model, "kmeans", b) for b in bits_list]
acc_lin = [evaluate_with_quant(model, "linear", b) for b in bits_list]

print(f"{'bits':>5} | {'K-Means':>9} | {'Linear':>9}")
for b, a1, a2 in zip(bits_list, acc_km, acc_lin):
    print(f"{b:>5} | {a1:>9.4f} | {a2:>9.4f}")
print(f"{'float':>5} | {base_acc:>9.4f} | {base_acc:>9.4f}")

plt.figure(figsize=(6.5, 4))
plt.axhline(base_acc, color="gray", ls=":", label=f"float32 ({base_acc:.3f})")
plt.plot(bits_list, acc_km,  "o-", color="#e0a020", label="K-Means")
plt.plot(bits_list, acc_lin, "s-", color="#22c55e", label="Linear (symmetric)")
plt.title("MNIST test accuracy vs bit-width")
plt.xlabel("bits per weight"); plt.ylabel("test accuracy")
plt.xticks(bits_list); plt.legend(); plt.grid(alpha=0.3)
plt.tight_layout(); plt.show()
```

## 마무리 & 더 해보기

- **관찰**: 낮은 비트(2~3비트)에서는 대개 K-Means(비균등)가 Linear(균등)보다 정확도를 잘 유지합니다. 가중치가 0 근처에 몰려 있기 때문입니다. 비트가 커지면(8비트) 둘 다 float에 근접합니다.
- 하지만 K-Means는 추론 시 코드북 디코드가 필요해 **저장만** 줄이고, Linear는 **정수 연산**까지 가능하다는 큰 차이가 있습니다 (블로그 글 참고).

**더 해볼 것**
1. `linear_quantize`의 `symmetric=False`(asymmetric)로 바꿔 정확도 비교
2. 활성값(activation)까지 양자화해 진짜 정수 추론 파이프라인 구현
3. 양자화 후 **코드북/스케일을 fine-tuning**해 정확도 되살리기 (Deep Compression의 재학습 단계)
4. Conv 층이 있는 모델(예: LeNet)로 확장
